# Why Vector Databases Are Needed for RAG Systems

Vector databases exist because traditional databases are **not optimized for nearest neighbor search over high-dimensional vectors**.

Similarity search works by computing the **distance between vectors in a high-dimensional space**. Modern embedding models generate vectors with **hundreds or thousands of dimensions**, which makes brute-force comparison across large datasets computationally expensive.

In embedding space:

- **Vectors close together** represent **semantically similar content**
- **Vectors far apart** represent **different concepts**

Because of this, vector databases are specifically designed to perform **efficient nearest neighbor search** across large collections of embeddings.

---

# Core Requirement of a Vector Database in RAG

The primary requirement for a vector database in a **Retrieval-Augmented Generation (RAG)** system is **fast similarity search**.

The database must efficiently handle:

- **Thousands or millions of vectors**
- **High-dimensional embeddings** (often 768–3072 dimensions)
- **Low-latency retrieval**

This enables the system to quickly retrieve the **most relevant documents or text chunks** when a user asks a question.

---

# High-Dimensional Embedding Space

Embedding models represent text as vectors in a **high-dimensional space**.

Example:

\[
x = [x_1, x_2, x_3, ..., x_n]
\]

Where:

- \(n\) = number of embedding dimensions  
- \(x_i\) = numerical value representing semantic features

Modern embedding models commonly produce vectors with:

- **768 dimensions**
- **1024 dimensions**
- **1536 dimensions**
- **3072 dimensions**

Searching through these vectors requires specialized **approximate nearest neighbor (ANN)** algorithms rather than traditional indexing methods.

---

# Key Capability: Nearest Neighbor Search

Vector databases optimize for **nearest neighbor retrieval**, which finds the vectors most similar to a query vector.

Similarity is typically measured using metrics such as:

### Cosine Similarity

\[
\text{similarity}(A,B) =
\frac{A \cdot B}{||A|| \; ||B||}
\]

### Euclidean Distance

\[
d(A,B) =
\sqrt{\sum_{i=1}^{n}(A_i - B_i)^2}
\]

These metrics allow the system to identify **semantically related documents** even when the wording differs.

### 1 Storing and Working with Embedding using FAISS

#### FAISS is well suited for RAG workflows which do not need persistence or metadata filtering.

````python
from IPython.display import Image, display

display(Image(filename="images/rwpc_0603.png", width=500, height=500))
````

````python
# Import FAISS library for efficient similarity search on vectors
import faiss

# Import NumPy for numerical operations and array handling
import numpy as np

# Import OpenAI client to access embedding models
from openai import OpenAI


# Example list of sample text strings (these act as our documents)
text_chunks = [
    "The sky is blue.",
    "The sun is shining.",
    "I love chocolate.",
    "Ice cream is delicious.",
    "Roses are red.",
    "Violets are blue.",
]


# Initialize the OpenAI client
# This will use your API key from environment variables
client = OpenAI()

# Select the embedding model
# text-embedding-3-small converts text into vector representations
model = "text-embedding-3-small"


# Function to generate embedding for a given text
def get_embedding(text):

    # Send the text to OpenAI embedding API
    response = client.embeddings.create(
        input=text,     # Text input to convert into embedding
        model=model     # Embedding model to use
    )

    # Extract the embedding vector from the API response
    return response.data[0].embedding


# Generate embeddings for each text chunk in the list
# List comprehension calls get_embedding() for every sentence
document_embeddings = np.array(
    [get_embedding(text) for text in text_chunks]
)


# Convert embeddings to float32 data type
# FAISS requires vectors in float32 format for indexing and search
document_embeddings = document_embeddings.astype("float32")


# Create a FAISS index for similarity search
# IndexFlatL2 means:
# - Flat index (no compression)
# - Uses L2 distance (Euclidean distance) for similarity
index = faiss.IndexFlatL2(document_embeddings.shape[1])


# Add all document embeddings to the FAISS index
# This builds the vector database so we can search later
index.add(document_embeddings)
````

````python
# Define the user query for which we want to find similar documents
query = "What color are violets?"

# Generate the embedding (vector representation) for the query text
# get_embedding() converts the query into a numerical vector
# np.array() converts it into a NumPy array
# astype("float32") ensures compatibility with FAISS (which requires float32)
query_embedding = np.array(get_embedding(query)).astype("float32")


# Perform the vector similarity search
# k represents the number of nearest (most similar) documents to retrieve
k = 5

# index.search() finds the k closest vectors in the FAISS index
# reshape(1, -1) converts the query vector into the required 2D format
# (1 query vector, embedding_dimension)
distances, indices = index.search(query_embedding.reshape(1, -1), k)


# Retrieve the original text documents corresponding to the returned indices
# indices[0] contains the positions of the most similar vectors
# We use these positions to fetch the actual text from text_chunks
retrieved_documents = [text_chunks[i] for i in indices[0]]
````

````python
print("Distances: ", distances)
print("Indices: ", indices)
print("Retrieved document: ", retrieved_documents)
````

# FAISS Overview

**FAISS** is a library for fast similarity search over dense vectors. It was released by Meta in 2017 and remains widely used because it is:

- Lightweight  
- Supports both CPU and GPU  
- Provides excellent performance for in-memory vector operations  

---

## How FAISS Works

FAISS organizes vectors into efficient **index structures** that enable fast **nearest neighbor search**.  

Common similarity metrics:

- **L2 (Euclidean distance):** Measures the straight-line distance between two points.  
- **Cosine similarity:** Measures the angle between vectors.  

Both metrics effectively identify **semantically similar content**.

---

## When to Use FAISS

Use FAISS when your workflow **does not require**:

- Persistence between runs  
- Metadata filtering  
- User-specific access controls  

FAISS excels at **temporary, single-use analysis tasks**:

- Example: Analyzing a contract to extract parties and dates  
- Example: Research scripts that process a corpus once and save results  

---

## When Not to Use FAISS

Avoid FAISS if you need:

- Data to survive application restarts  
- Multiple users with different access permissions  
- Filtering by metadata fields before searching  

> FAISS stores everything in memory or in a simple file format with **no built-in query language or access control**.

---

## Tradeoffs

**FAISS** focuses on **simplicity and speed**:

- Fastest pure vector search performance  
- Minimal code required  
- No database capabilities  

You must handle **persistence, access control, and metadata filtering** in your application layer.

---

## Comparison with Other Solutions

| Feature / Tool | FAISS | ChromaDB | PostgreSQL + pgvector |
|----------------|-------|----------|----------------------|
| Persistence | ❌ In-memory (can serialize) | ✅ Lightweight | ✅ Full SQL, transactional |
| Metadata Filtering | ❌ | ✅ Basic | ✅ Full SQL support |
| Access Control | ❌ | ❌ | ✅ Role-based |
| Use Case | Batch jobs, offline indexing | Prototypes needing simple persistence | Production systems requiring reliability |

## Production Recommendations

- For **persistence, metadata filtering, multi-user access**, use:  
  - **PostgreSQL with pgvector** (if you already use PostgreSQL)

These solutions support **role-based access control** and are available as **managed services** on cloud platforms.

#### 2 Storing and Working with Embeddings in ChromaDB

##### ChromaDB is a lightweight vector database designed for quick prototyping and small projects. It provides persistence and metadata handling while remaining simple to set up and use.

ChromaDB can store your data in two ways:

Ephemeral client: stores data in memory, which means the data is lost when the application stops.

Persistent client: saves data to disk, so you can reload it when restarting your app.

ChromaDB has built-in functions that can handle text processing and embedding generation automatically. For prototypes, these built-in features speed up development. For production systems you plan to migrate later, generate embeddings in your own code and use ChromaDB only to store and search them. This makes switching to PostgreSQL or Pinecone straightforward.

1. Load the text, chunk it, and generate embeddings using an embedding model.

2. Insert those embeddings into ChromaDB. If persistence is enabled, ChromaDB also stores them on disk.

````python
# Import the ChromaDB library for persistent vector storage
import chromadb

# ------------------------------
# VECTOR STORE SETTINGS
# ------------------------------
# Path where vector embeddings will be stored on disk
VECTOR_STORE_PATH = r"../02_Data/00_Vector_Store"

# Name of the collection (like a table in a database)
COLLECTION_NAME = "my_collection"

# ------------------------------
# INITIALIZE CHROMA CLIENT
# ------------------------------
# Create a persistent Chroma client, storing data at VECTOR_STORE_PATH
chroma_client = chromadb.PersistentClient(path=VECTOR_STORE_PATH)

# Get an existing collection or create a new one if it doesn't exist
collection = chroma_client.get_or_create_collection(name=COLLECTION_NAME)

# ------------------------------
# IMPORT OPENAI CLIENT
# ------------------------------
from openai import OpenAI

# Sample text chunks to be embedded and stored in the vector store
text_chunks = [
    "The sky is blue.",
    "The sun is shining.",
    "I love chocolate.",
    "Ice cream is delicious.",
    "Roses are red.",
    "Violets are blue.",
]

# ------------------------------
# FUNCTION TO GENERATE EMBEDDINGS
# ------------------------------
# Initialize the OpenAI client
client = OpenAI()

# Specify the embeddings model
model = "text-embedding-3-small"

# Function to generate vector embeddings for a given text
def get_embedding(text):
    response = client.embeddings.create(input=text, model=model)
    return response.data[0].embedding  # return the embedding vector

# For compatibility with later code (could use embeddings_model directly)
embeddings_model = client  

# Generate embeddings for all text chunks
embeddings_list = [get_embedding(text) for text in text_chunks]

# ------------------------------
# ADD DATA TO THE CHROMA COLLECTION
# ------------------------------
collection.add(
    embeddings=embeddings_list,  # list of embeddings
    documents=text_chunks,       # corresponding text chunks
    ids=[str(i) for i in range(len(text_chunks))],  # string IDs for each chunk
)

# ------------------------------
# QUERY THE COLLECTION
# ------------------------------
# Example query to find similar text chunks
query = "What is the color of the sky?"

# Generate embedding for the query
query_embedding = get_embedding(query)

# Perform similarity search in the collection and return top 3 results
results = collection.query(query_embeddings=[query_embedding], n_results=3)

# Print results for demonstration
print(results)
````

**ChromaDB** works by storing **vectors with their metadata** in a simple embedded database. It **persists data to disk automatically**, so you can restart your application without rebuilding the index.

Use ChromaDB when you need:

- **Persistence** between application runs  
- **Basic metadata filtering**  
- Minimal setup for quick development  

Ideal use cases:

Avoid ChromaDB when:

- Your dataset grows beyond a few hundred thousand vectors  
- You need **role-based access control**  
- You require **SQL joins with application data**  
- Query latency must be **under 10ms at scale**  

Limitations:

- Lacks advanced security features  
- No sophisticated query optimization  
- Cannot scale horizontally  

---

## Tradeoffs

The main tradeoff with ChromaDB is **ease of use versus scalability**:

- **Pros:** Minimal dependencies, start coding immediately  
- **Cons:** Performance degrades as dataset grows  

ChromaDB provides a **middle ground** between:

- **FAISS:** Can be saved to disk but lacks database features  
- **PostgreSQL + pgvector:** Full database features  

---

## Key Differences

- **ChromaDB vs FAISS:** Adds **persistence** and **metadata filtering**, allowing app restarts without rebuilding the index.  
- **ChromaDB vs PostgreSQL + pgvector:** Simpler setup but lacks **SQL, transactions, and production-grade access control**.  

> When your dataset grows beyond a few hundred thousand vectors or you need **role-based access and SQL joins**, migrate to **PostgreSQL with pgvector** or a managed vector database like **Pinecone**.

# PostgreSQL with pgvector Overview

**PostgreSQL with pgvector** stores **embeddings as native vector columns** alongside your application data.  
The **pgvector extension** adds:

- Vector data types  
- Similarity search operators  

directly to SQL, allowing you to query embeddings with **familiar PostgreSQL syntax**.  

This enables you to **join user tables, document metadata, and embeddings** in a single query without managing separate systems.

---

## When to Use PostgreSQL + pgvector

Use PostgreSQL with pgvector when:

- Your application **already uses PostgreSQL** for structured data  
- You need to **filter embeddings by metadata** (e.g., user permissions, dates, categories) before similarity search  
- You want to **avoid managing a separate vector database**  

PostgreSQL is well-suited for **production RAG systems** with up to **tens of millions of vectors**, especially when you need:

- **ACID transactions**  
- **Role-based access control**  

---

## When Not to Use PostgreSQL + pgvector

Avoid PostgreSQL when:

- You need to search **hundreds of millions or billions of vectors** with **single-digit millisecond latency**  
  - Use purpose-built vector databases like **Pinecone** or **Weaviate** instead  
- Your workload is **purely vector search** with no metadata filtering  
  - **FAISS** or specialized vector databases will be faster  

---

## Tradeoffs

The main tradeoff with PostgreSQL is **integration vs raw performance**:

- **Pros**:  
  - Unified data management  
  - SQL joins  
  - Transactional guarantees  

- **Cons**:  
  - Specialized vector databases are faster for **pure vector workloads at very large scale**  

For most **RAG applications**, the integration benefits outweigh the performance difference.

---

**Summary**:

#### 4 Perform similarity search on top of the just create PostgreSQL table

# Similarity Search in PostgreSQL

PostgreSQL similarity search works by **computing distance metrics** between a query embedding and stored embeddings using SQL operators.

- **Distance operators**:
  - `<=>` : Cosine distance
  - `<->` : Euclidean distance
  - `<#>` : Inner product

PostgreSQL executes these calculations either by:

- Scanning all rows (**full table scan**)  
- Using **vector indexes** (HNSW or IVF) to narrow the search space for faster retrieval

---

## Combining Similarity Search with SQL Filtering

The power of PostgreSQL similarity search comes from **combining it with SQL filtering**:

- Use `WHERE` clauses to filter by metadata **before computing similarity**.  
- This reduces the search space and improves relevance.  

**Examples of filters**:

- Documents accessible by the current user  
- Published after a certain date  
- Tagged with specific categories  

**Use cases**:

- Legal RAG system: search only contracts the user has permission to view  
- News RAG system: search only articles from the last 30 days  

```sql
SELECT *
FROM documents
WHERE user_id = 42 AND published_date > '2026-01-01'
ORDER BY embedding <=> '<query_embedding>' ASC
LIMIT 10;

````python
# ------------------------------
# IMPORT OPENAI CLIENT
# ------------------------------
from openai import OpenAI

# ------------------------------
# INITIALIZE OPENAI EMBEDDINGS MODEL
# ------------------------------
client = OpenAI()               # Connect to OpenAI API
model = "text-embedding-3-small"  # Choose embedding model

# ------------------------------
# EXAMPLE QUERY TEXT
# ------------------------------
text_chunk = "Sweets are delicious."  # Query text we want to find similar chunks for

# ------------------------------
# GENERATE EMBEDDING FOR QUERY
# ------------------------------
response = client.embeddings.create(input=text_chunk, model=model)
embeded_query = response.data[0].embedding  # 1536-dimensional embedding vector

# ------------------------------
# CREATE CURSOR
# ------------------------------
cur = conn.cursor()

# ------------------------------
# RUN VECTOR SIMILARITY QUERY USING PGVECTOR
# ------------------------------
# '<=>' operator computes the cosine distance between vectors
# 1 - (embedding <=> query_vector) gives cosine similarity
# Results are ordered by similarity descending (most similar first)
cur.execute(
    f"""
    SELECT 1 - (embedding <=> '{embeded_query}') AS cosine_similarity, *
    FROM embeddings
    ORDER BY 1 - (embedding <=> '{embeded_query}') DESC
    LIMIT 20;
"""
)

# ------------------------------
# FETCH AND INSPECT RESULTS
# ------------------------------
results = cur.fetchall()

for row in results:
    print(row)

# ------------------------------
# CLEAN UP
# ------------------------------
cur.close()
conn.close()
````

### 6 Perform a vector search using HNSW and IVFLL index

# PostgreSQL Vector Indexing: HNSW and IVF

For tables under **100,000 rows**, full table scans are often **fast enough**.  
- In this case, **skip indexing**.  

For tables above **1 million rows**, indexing becomes **essential** to maintain fast query performance.

---

## Choosing an Index Type

- **HNSW (Hierarchical Navigable Small World)**  
  - Best when **query speed is critical**  
  - Requires **sufficient memory** (stores more data in RAM)  
  - Typically provides **best performance** for millions of vectors  

- **IVF (Inverted File Index)**  
  - Use when **memory is constrained**  
  - Faster **index building**, slightly slower queries  
  - Good tradeoff if you need quick setup on large datasets  

---

## Similarity Metric Operators

When creating vector indexes, choose the operator based on your similarity metric:

| Metric                  | Index Operator          |
|-------------------------|------------------------|
| Cosine similarity       | `vector_cosine_ops`    |
| Inner product           | `vector_ip_ops`        |
| Euclidean distance (L2) | `vector_l2_ops`        |

---

## HNSW Hyperparameters

- **m**: Maximum connections per layer in the HNSW graph  
  - Higher values → denser graph → faster search  
  - Increases **index build time and memory usage**  
  - Typical range: **16–64**  

- **ef_construction**: Size of the dynamic candidate list during index construction  
  - Higher values → better index quality  

- **ef_search**: Number of candidates considered during search  
  - Higher values → more accurate results  
  - Higher values → slower queries  

> Understanding these parameters helps you **tune HNSW** for your dataset and workload.

---

## IVF Index Overview

- **IVF** = **Inverted File**  
- Divides vector space into **partitions**, each with a **centroid**  
- Each vector belongs to **only one partition**  
- During a search:  
  1. The system compares the query to **centroids**  
  2. Searches only in the **closest partitions**  
- Trades **slightly lower accuracy** for **faster query speed** and lower memory usage  

---

### ✅ Summary

- **Small datasets (<100k rows)** → full scan is fine  
- **Medium to large datasets (>1M rows)** → use vector indexes  
- **HNSW** → fast, accurate, memory-intensive  
- **IVF** → memory-efficient, slightly slower, easier to build  
- Tune **HNSW hyperparameters** (`m`, `ef_construction`, `ef_search`) for best performance  
- Choose **index operator** according to your similarity metric

````python
# ------------------------------
# DEFINE USER QUERY
# ------------------------------
query = "I am looking for a job as a data scientist in Berlin."

# ------------------------------
# INITIALIZE OPENAI CLIENT
# ------------------------------
client = OpenAI()
model = "text-embedding-3-small"

# ------------------------------
# GENERATE EMBEDDING FOR QUERY
# ------------------------------
# Convert the query text into a 1536-dimensional vector
response = client.embeddings.create(input=query, model=model)
query_embedding = response.data[0].embedding

# ------------------------------
# CREATE SQL FOR VECTOR SIMILARITY SEARCH
# ------------------------------
# '<=>' operator computes cosine distance
# 1 - (distance) converts it to cosine similarity
# ORDER BY DESC ensures most similar entries come first
# LIMIT 20 returns top 20 results
# EXPLAIN ANALYZE is used to measure query execution time
full_search_sql = f"""
    EXPLAIN ANALYZE
    SELECT 1 - (embedding <=> '{str(query_embedding)}') AS cosine_similarity,
           *
    FROM job_description_table
    ORDER BY 1 - (embedding <=> '{str(query_embedding)}') DESC
    LIMIT 20;
"""

# ------------------------------
# EXECUTE QUERY
# ------------------------------
cur.execute(full_search_sql)

# Fetch results (execution plan + estimated times)
results_full_search = cur.fetchall()

# Print results to inspect query performance
for row in results_full_search:
    print(row)
````

### 7 Hybrid search using PostgreSQL

# Hybrid Search in PostgreSQL

Hybrid search combines **keyword search (full-text)** and **semantic search (vector similarity)** to retrieve documents that:

- Contain the **exact query terms**, or  
- Express **similar concepts**  

---

## How It Works

1. **Keyword search**:  
   - Uses PostgreSQL’s `tsvector` type and `ts_rank` function  
   - Identifies documents containing query terms  

2. **Semantic search**:  
   - Uses cosine similarity (`1 - (embedding <=> query_embedding)`)  
   - Finds documents **close in meaning** to the query  

3. **Hybrid scoring**:  
   - Compute both text and vector scores  
   - Combine them with **weights** to rank results  
   - Example: `hybrid_score = 0.5 * text_score + 0.5 * vector_score`

---

## When to Use Hybrid Search

Use hybrid search when **exact keyword matches matter**, such as:

- **Legal documents**: searching for `"force majeure"` or `"indemnification"`  
- **Technical documentation**: searching for **error codes**, **API names**  
- **Medical records**: searching for **drug names** or **diagnosis codes**  
- Queries containing **proper nouns, acronyms, or domain-specific terminology**  

---

## When Not to Use Hybrid Search

Avoid hybrid search when:

- **Semantic meaning alone is sufficient**  
  - Example: conversational customer support tickets, general knowledge articles  
- **Keyword index is poorly tuned**  
  - Poor keyword results will reduce overall retrieval quality  

---

## Trade-offs

- **Complexity vs Robustness**:  
  - Must maintain **both a text search index** and **a vector index**  
  - Tune **two sets of parameters**  
  - Determine **appropriate weights** for combining scores  

- **Weight tuning tips**:  
  - Start with **equal weights (0.5 text, 0.5 vector)**  
  - Increase vector weight to **0.7** if relevant results are missing due to lack of exact terms  
  - Increase text weight to **0.7** if results drift too far from the literal query  

- Domain-specific defaults:  
  - Legal / technical → higher **text weight**  
  - Conversational / general → higher **vector weight**

---

## Comparison With Other Approaches

| Approach                        | Exact Terms | Semantic Meaning | Metadata Filtering | Notes |
|---------------------------------|------------|----------------|-----------------|-------|
| Pure semantic search             | ❌         | ✅             | ❌               | Fast, meaning-based retrieval |
| Pure keyword search              | ✅         | ❌             | ❌               | Finds only literal matches |
| SQL filtering + vector search    | ❌         | ✅             | ✅               | Filters by metadata before semantic search |
| **Hybrid search**                | ✅         | ✅             | ❌               | Combines literal keyword matches with semantic similarity |

**Rule of thumb**:

- **Hybrid search** → when **both exact terms and meaning matter**  
- **SQL filtering** → when you need **metadata-based restrictions**  
- **Pure semantic search** → when **meaning alone is sufficient**

````python
# ------------------------------
# IMPORT LIBRARIES
# ------------------------------
import psycopg2
from psycopg2 import Error
from openai import OpenAI

# ------------------------------
# DEFINE QUERY
# ------------------------------
query = "I am looking for a job as a data scientist in Berlin."

# ------------------------------
# INITIALIZE OPENAI CLIENT
# ------------------------------
client = OpenAI()
model = "text-embedding-3-small"

# Generate embedding for the query text
response = client.embeddings.create(input=query, model=model)
query_embedding = response.data[0].embedding  # 1536-dim vector

# ------------------------------
# HYBRID SEARCH SQL
# ------------------------------
# Combines full-text search (tsv / ts_rank) and vector similarity (embedding)
# Steps:
# 1. Compute text score using ts_rank (PostgreSQL full-text search ranking)
# 2. Compute vector score using cosine similarity (1 - embedding <=> query_embedding)
# 3. Compute hybrid_score = weighted sum of text_score and vector_score (0.5 + 0.5)
# 4. Filter rows using full-text search (tsv @@ plainto_tsquery)
# 5. Order by hybrid_score descending and return top 10
hybrid_search_query = f"""
    WITH ranked_docs AS (
        SELECT
            id,
            text_chunk AS text,
            ts_rank(tsv, plainto_tsquery('PostgreSQL')) AS text_score,
            1 - (embedding <=> '{str(query_embedding)}') AS vector_score,
            ts_rank(tsv, plainto_tsquery('PostgreSQL'))
            * 0.5 + (1 - (embedding <=> '{str(query_embedding)}')) * 0.5
            AS hybrid_score
        FROM test_embedding_table
        -- Filter for relevant text first
        WHERE tsv @@ plainto_tsquery('PostgreSQL')
        ORDER BY hybrid_score DESC
        LIMIT 10
    )
    SELECT * FROM ranked_docs;
"""

# ------------------------------
# EXECUTE HYBRID SEARCH
# ------------------------------
cur.execute(hybrid_search_query)

# Fetch top results
results = cur.fetchall()

# ------------------------------
# PRINT RESULTS
# ------------------------------
for row in results:
    print(row)
````
